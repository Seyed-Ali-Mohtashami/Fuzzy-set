```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0.0, 1.01, 0.01,dtype=float)
plt.figure(figsize=(10,10))
y1=np.array([0.0]*x.size)
y2=np.array([0.0]*x.size)
y3=np.array([0.0]*x.size)
y4=np.array([0.0]*x.size)
y5=np.array([0.0]*x.size)

#**********************************
s=-0.95
for i in range(x.size):    
    a=x[i]
    N=(float)((1-a)/(1+s*a))
    y1[i]=N
#**********************************
s=-0.7
for i in range(x.size):    
    a=x[i]
    N=(float)((1-a)/(1+s*a))
    y2[i]=N
#**********************************
s=-0
for i in range(x.size):    
    a=x[i]
    N=(float)((1-a)/(1+s*a))
    y3[i]=N
#**********************************
s=2
for i in range(x.size):    
    a=x[i]
    N=(float)((1-a)/(1+s*a))
    y4[i]=N
#**********************************
s=20
for i in range(x.size):    
    a=x[i]
    N=(float)((1-a)/(1+s*a))
    y5[i]=N

    
plt.subplot(221)
plt.plot(x,y1,x,y2,x,y3,x,y4,x,y5)
plt.xlabel('X')
plt.ylabel('N(a)')
plt.xticks([0,0.5,1])
plt.ylabel([0,0.2,0.4,0.6,0.8,1])
plt.text(0.85,0.8,'s=-0.95')
plt.text(0.7,0.6,'s=-0.70')
plt.text(0.5,0.55,'s=0.00')
plt.text(0.3,0.45,'s=2.00')
plt.text(0.2,0.2,'s=20.0')
plt.title('(a) sugeno complements')


#****************************************************************************
#**********************************
w=3
for i in range(x.size):    
    a=x[i]
    N=(float)(1-a**w)**(1/w)
    y1[i]=N
#**********************************
w=1.5
for i in range(x.size):    
    a=x[i]
    N=(float)(1-a**w)**(1/w)
    y2[i]=N
#**********************************
w=1
for i in range(x.size):    
    a=x[i]
    N=(float)(1-a**w)**(1/w)
    y3[i]=N
#**********************************
w=0.7
for i in range(x.size):    
    a=x[i]
    N=(float)(1-a**w)**(1/w)
    y4[i]=N
#**********************************
w=0.4
for i in range(x.size):    
    a=x[i]
    N=(float)(1-a**w)**(1/w)
    y5[i]=N

    
plt.subplot(222)
plt.plot(x,y1,x,y2,x,y3,x,y4,x,y5)
plt.xlabel('X')
plt.ylabel('N(a)')
plt.xticks([0,0.5,1])
plt.ylabel([0,0.2,0.4,0.6,0.8,1])
plt.text(0.9,0.8,'w=3')
plt.text(0.7,0.6,'w=1.5')
plt.text(0.5,0.55,'w=1')
plt.text(0.3,0.45,'w=0.7')
plt.text(0.2,0.2,'w=0.4')
plt.title('(b) yager complements')

plt.show()
```
