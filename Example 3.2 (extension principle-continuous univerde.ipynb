```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzzy
#---------------------------- mx , x ----------------------------
x=np.arange(-3,3,0.01)
#print(x)
mx=fuzzy.gbellmf(x,1.5,2,0.5)
mx=np.round(mx,2)
#print(mx)
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.plot(x,mx)
plt.xticks(np.arange(-3,4,1))
plt.yticks(np.arange(0,1.4,0.2))
plt.xlabel('X')
plt.ylabel('Membership graides')
plt.text(0.5,1.1,'A')
plt.axhline(0.5,color='gray',linewidth='0.5')
plt.axvline(2,color='gray',linewidth='0.5')
plt.axvline(-1,color='gray',linewidth='0.5')

#------------------------------ f(x) -----------------------------
def f(xx):
    yy=np.zeros(xx.size)
    for i in range(xx.size):
        if xx[i]>= 0:
            yy[i]=(xx[i]-1)**2-1
        else:
            yy[i]=xx[i]
    return yy
#                             show f(x)
y=f(x)
y=np.round(y,2)
plt.subplot(222)
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.text(-0.5,2,'y=f(x)')
plt.axvline(2,color='gray',linewidth='0.5')
plt.axvline(-1,color='gray',linewidth='0.5')
plt.axhline(0,color='gray',linewidth='0.5')
plt.axhline(-1,color='gray',linewidth='0.5')
#---------------------------- my , y ----------------------------
my=mx.copy()

for i in range(my.size):
    fa=np.where(y==y[i])
    f=fa[0]
    value=my[i]
    for j in range(f.size):        
        if my[f[j]]> value:            
            value=my[f[j]]
    my[i]=value

plt.subplot(223)
plt.plot(y,my)
plt.axvline(0,color='gray',linewidth='0.5')
plt.axvline(-1,color='gray',linewidth='0.5')
plt.text(1,0.8,'B')
#-------------------------------------------------------------------
plt.show()
```
