```python
import skfuzzy as fuzzy
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,90,1)
y_young=fuzzy.gbellmf(x,30,5,0)
y_middle=fuzzy.gbellmf(x,15,3,45)
y_old=fuzzy.gbellmf(x,30,5,90)

plt.figure(figsize=(30,10))
plt.plot(x,y_young,x,y_middle,x,y_old)

plt.text(5,1.1,'young',fontsize=30)
plt.text(40,1.1,'middel',fontsize=30)
plt.text(75,1.1,'old',fontsize=30)

font1 = {'family':'serif','color':'blue','size':30}
font2 = {'family':'serif','color':'darkred','size':15}
plt.xlabel('X=age',fontdict=font1)
plt.ylabel('Membership Grades',fontdict=font1)


plt.xticks(np.arange(0,91,10),fontsize=20)
plt.yticks(np.arange(0,1.3,0.2),fontsize=20)


plt.show()
```
