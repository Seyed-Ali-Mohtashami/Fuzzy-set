```python
import numpy as np
import matplotlib.pyplot as plt


a=np.arange(0,1.1,0.1)
b=np.arange(0,1.1,0.1)
a,b=np.meshgrid(a,b)

fig=plt.figure(figsize=(8,8))
#-------------------------- maximum -----------------------------
z=a.copy()
for i in range(11):
    for j in range(11):
        if a[i][j]>= b[i][j]:
            z[i][j]=a[i][j]
        else:
            z[i][j]=b[i][j]
ax=fig.add_subplot(221,projection='3d')
wf=ax.plot_wireframe(a,b,z)
ax.set_title('(a) max')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xticks([0,0.5,1])
ax.set_yticks([0,0.5,1])
ax.set_zticks([0,0.5,1])
#-------------------------- algebric sum -----------------------------
z=a.copy()
for i in range(11):
    for j in range(11):         
        z[i][j]=a[i][j]+b[i][j]-a[i][j]*b[i][j]
ax=fig.add_subplot(222,projection='3d')
wf=ax.plot_wireframe(a,b,z)
ax.set_title('(b) algebraic sum')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xticks([0,0.5,1])
ax.set_yticks([0,0.5,1])
ax.set_zticks([0,0.5,1])
#-------------------------- bounded sum -----------------------------
z=a.copy()
for i in range(11):
    for j in range(11):   
        c=a[i][j]+b[i][j]
        if  c >=1 :
            z[i][j]=1
        else:
            z[i][j]=c
ax=fig.add_subplot(223,projection='3d')
wf=ax.plot_wireframe(a,b,z)
ax.set_title('(c) bounded sum')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xticks([0,0.5,1])
ax.set_yticks([0,0.5,1])
ax.set_zticks([0,0.5,1])
#-------------------------- drastic sum -----------------------------
z=a.copy()
for i in range(11):
    for j in range(11):           
        if  b[i][j] ==0:
            z[i][j]=a[i][j]
        elif a[i][j]==0:
            z[i][j]=b[i][j]
        else:
            z[i][j]=1
ax=fig.add_subplot(224,projection='3d')
wf=ax.plot_wireframe(a,b,z)
ax.set_title('(d) drastic sum')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xticks([0,0.5,1])
ax.set_yticks([0,0.5,1])
ax.set_zticks([0,0.5,1])

plt.show()
```
