```python
import skfuzzy as fuzzy
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(1,100,1)


plt.figure(figsize=(10,10))
plt.subplots_adjust(wspace=0.5,hspace=0.5)

#----------------------------------- trianguler ----------------------------------- 
plt.subplot(221)
y_tri=fuzzy.trimf(x,(20,60,80))
plt.plot(x,y_tri)
plt.title('(a) trianguler MF')
plt.ylabel('Membership Grades')

#---------------------------------- trapezoidal -----------------------------------
plt.subplot(222)
y_trap=fuzzy.trapmf(x,[10,20,60,95])
plt.plot(x,y_trap)
plt.title('(b) trapezoidal MF')
plt.ylabel('Membership Grades')

#------------------------------------ gaussian ------------------------------------
plt.subplot(223)
y_gaus=fuzzy.gaussmf(x,50,20)
plt.plot(x,y_gaus)
plt.title('(c) guassian MF')
plt.ylabel('Membership Grades')

#-------------------------------------- bell --------------------------------------
plt.subplot(224)
y_bell=fuzzy.gbellmf(x,20,4,50)
plt.plot(x,y_bell)
plt.title('(d) bell MF')
plt.ylabel('Membership Grades')


plt.show()
```
