```python
import skfuzzy as fuzzy
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-10,10,0.1)

plt.figure(figsize=(10,10))

#----------------------------------- change a ----------------------------------- 
b=2
c=0
y1=fuzzy.gbellmf(x,2,b,c)
y2=fuzzy.gbellmf(x,4,b,c)
y3=fuzzy.gbellmf(x,6,b,c)
plt.subplot(221)
plt.title('(a) changing \'a\'')
plt.plot(x,y1,x,y2,x,y3)

#----------------------------------- change b ----------------------------------- 
a=5
c=0
y1=fuzzy.gbellmf(x,a,1,c)
y2=fuzzy.gbellmf(x,a,2,c)
y3=fuzzy.gbellmf(x,a,4,c)
plt.subplot(222)
plt.title('(b) changing \'b\'')
plt.plot(x,y1,x,y2,x,y3)

#----------------------------------- change c ----------------------------------- 
a=5
b=2
y1=fuzzy.gbellmf(x,a,b,-5)
y2=fuzzy.gbellmf(x,a,b,0)
y3=fuzzy.gbellmf(x,a,b,+5)
plt.subplot(223)
plt.title('(c) changing \'c\'')
plt.plot(x,y1,x,y2,x,y3)

#-------------------------------- change a and b --------------------------------- 
c=0
y1=fuzzy.gbellmf(x,4,4,c)
y2=fuzzy.gbellmf(x,6,6,c)
y3=fuzzy.gbellmf(x,8,8,c)
plt.subplot(224)
plt.title('(d) changing \'a\' and \'b\'')
plt.plot(x,y1,x,y2,x,y3)

plt.show()
```
