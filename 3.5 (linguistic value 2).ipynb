```python
import skfuzzy as fuzzy
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,90,1)
y_young=fuzzy.gbellmf(x,30,5,0)
#y_middle=fuzzy.gbellmf(x,15,3,45)
y_old=fuzzy.gbellmf(x,30,5,90)

plt.figure(figsize=(10,10))

#--------------------------------------- young - middle -----------------------------------------
plt.subplot(511)
plt.plot(x,y_young,x,y_old)

plt.text(5,1.1,'young',fontsize=10)
plt.text(75,1.1,'old',fontsize=10)
plt.yticks([])
plt.xticks([])
#--------------------------------------- young - middle -----------------------------------------


#--------------------------------- not young and not middle -------------------------------------
plt.subplot(512)

y_not_young=1-y_young
y_not_old=1-y_old

y_not_yound_and_not_old=np.zeros(x.size)

for i in range(x.size):
    y_not_yound_and_not_old[i]=min(y_not_young[i],y_not_old[i])

plt.plot(x,y_not_yound_and_not_old)
plt.text(35,0.4,'not young and not old',fontsize=10)
plt.yticks([])
plt.xticks([])
#--------------------------------- not young and not middle -------------------------------------


#---------------------------------- young but not too young -------------------------------------
plt.subplot(513)

y_too_young=y_young ** 2
y_not_too_young=1-y_too_young
y_yount_but_not_too_young=np.zeros(x.size)

for i in range(x.size):
    y_yount_but_not_too_young[i]=min(y_young[i],y_not_too_young[i])

plt.plot(x,y_yount_but_not_too_young)
plt.text(35,0.4,'young but not too young',fontsize=10)
plt.yticks([])
plt.xticks([])
#---------------------------------- young but not too young -------------------------------------


#-------------------------------------- more or less old ----------------------------------------
plt.subplot(514)

y_less_old=y_old ** 0.5

plt.plot(x,y_less_old)
plt.text(35,0.4,'more or less old',fontsize=10)
plt.yticks([])
plt.xticks([])
#-------------------------------------- more or less old ----------------------------------------


#---------------------------------------- extermely old -----------------------------------------
plt.subplot(515)

y_extermely_old=y_old ** 8

plt.plot(x,y_extermely_old)
plt.text(35,0.4,'extermely old',fontsize=10)
plt.yticks([])
plt.xticks([])
#---------------------------------------- extermely old -----------------------------------------


plt.show()
```
