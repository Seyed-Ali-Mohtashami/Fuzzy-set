```python
import numpy as np
import skfuzzy as fuzzy
import matplotlib.pyplot as plt

X=np.array([3,4,5])
Y=np.array([3,4,5,6,7])

y,x=np.meshgrid(Y,X)

def f_r(xx,yy):
        rr=np.zeros(xx.shape)
        r,c=xx.shape
        for i in range(r):
            for j in range(c):
                if xx[i,j] >= y[i,j]:
                    rr[i,j]=0
                else :
                    rr[i,j]=(yy[i,j]-xx[i,j])/(yy[i,j]+xx[i,j]+2)                    
        return rr
r=f_r(x,y)
print(r.round(2))

fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(111,projection='3d')
m=ax.scatter(x,y,r)
#m=ax.plot_wireframe(x,y,r)
m=ax.plot_surface(x,y,r,alpha=0.5)

ax.set_title('binary fuzzy relation')
ax.set_xticks(X)
ax.set_yticks(Y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('R')

plt.show()
```
