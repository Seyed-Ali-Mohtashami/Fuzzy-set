```python
import skfuzzy as fuzzy
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-10,10,0.1)
plt.figure(figsize=(10,10))
#--------------------------------------y1 and y2----------------------------------- 
plt.subplot(221)
plt.title('(a) sig(x,-5,1) ; sig(x,5,2)')
y1=fuzzy.sigmf(x,-5,1)
y2=fuzzy.sigmf(x,5,2)
plt.plot(x,y1,x,y2)

#-------------------------------------- |y1 - y2| --------------------------------- 
plt.subplot(222)
plt.title('(b) |y1 - y2|')

#y_new=y1-y2

# manual set---> y_new= y1 or (not y2)
y_new=np.arange(y1.size,dtype=float)
for i in range(y1.size):
    newy=1-y2[i]
    if y1[i] < newy:
        y_new[i]=y1[i]
    else:
        y_new[i]=newy
plt.plot(x,y_new)
  
#--------------------------------------y1 and y3----------------------------------- 
plt.subplot(223)
plt.title('(c) sig(x,-5,1) ; sig(x,5,-2)')
y1=fuzzy.sigmf(x,-5,1)
y3=fuzzy.sigmf(x,5,-2)
plt.plot(x,y1,x,y3)  

#-------------------------------------- y1 * y2 --------------------------------- 
plt.subplot(224)
plt.title('(d) y1 * y2')

#y_new=y1*y2

# manual set---> y_new= y1 and y3
y_new=np.arange(y1.size,dtype=float)
for i in range(y1.size):    
    if y1[i] < y3[i]:
        y_new[i]=y1[i]
    else:
        y_new[i]=y3[i]
        
plt.plot(x,y_new)
  
    
plt.show()
```
