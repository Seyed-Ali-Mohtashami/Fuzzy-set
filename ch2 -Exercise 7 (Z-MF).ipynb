```python
import numpy as np
import matplotlib.pyplot as plt

def S(x,l,r):    
    y=np.zeros(x.size)
    for i in range(x.size):
        if x[i]<=l:
            y[i]=0
        elif x[i]<= ((l+r)/2):
            y[i]=(2*(((x[i]-l)/(r-l))**2))
        elif x[i] <=r:
            y[i]=(1-(2*((r-x[i])/(r-l))**2))
        else:
            y[i]=(1)
    return y


def Z(x,l,r):
    return(1-S(x,l,r))
           
X=np.arange(0,101,1)
l=20
r=50
y1=Z(X,l,r)
l=50
r=80
y2=Z(X,l,r)
plt.xlabel('X')
plt.title('Z MF')
plt.text(15,0.6,'l=20,r=50')
plt.text(75,0.6,'l=50,r=80')
plt.plot(X,y1,X,y2)
plt.show()

print('crossover point of Z(x,l,r) is : x=(l+r)/2')
```
