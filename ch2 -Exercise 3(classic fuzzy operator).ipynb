```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzzy

X=np.arange(1,21,1)
print('X='+str(X))
#---------------------------- A(divisible 3) ----------------------------
A=[]
for i in range(X.size):
    if X[i] % 2 ==0:
            A.append(X[i])
print('A='+str(A))
#---------------------------- B(divisible 6) ----------------------------
B=[]
for i in range(X.size):
    if X[i] % 3 ==0:
            B.append(X[i])
print('B='+str(B))
#---------------------------- B(divisible 5) ----------------------------
C=[]
for i in range(X.size):
    if X[i] % 5 ==0:
            C.append(X[i])
print('C='+str(C))
print()
#---------------------------- ¬A ----------------------------
def f_not_A(x,a):  
    not_a=[]
    for i in range(x.size):
        if a.count(x[i])>0:
            continue
        else:
            not_a.append(x[i])
    return not_a
not_A=f_not_A(X,A)
print('¬A ='+str(not_A)) 
print()
#---------------------------- A ∩ ¬A ----------------------------
def f_con(x,a,b):
    z=[]
    for i in range(x.size):
        if a.count(x[i])>0 and b.count(x[i])>0:
            z.append(x[i])
    return z
con_A_notA=f_con(X,A,not_A)
print('A ∩ ¬A = ѳ :',con_A_notA)   
print()
#---------------------------- A ∪ ¬A  ----------------------------
def f_disj(x,a,b):
    z=[]
    for i in range(x.size):
        if a.count(x[i])>0 or b.count(x[i])>0:
            z.append(x[i])
    return z
disj_A_notA=f_disj(X,A,not_A)
print('A ∪ ¬A = X:',disj_A_notA)   
print()
#---------------------------- A ∩ A ----------------------------
con_A_A=f_con(X,A,A)
print('A ∩ A = A: ',con_A_A)   
print()
#---------------------------- A ∪ A  ----------------------------
disj_A_A=f_disj(X,A,A)
print('A ∪ A = A: ',disj_A_A)   
print()
#---------------------------- ¬(¬A) = A----------------------------
not_not_A=f_not_A(X,not_A)
print('¬(¬A)= A : '+str(not_not_A))
print()
#---------------------------- A ∩ B = B ∩ A ----------------------------
con_A_B=f_con(X,A,B)
print('A ∩ B = '+str(con_A_B))
con_B_A=f_con(X,B,A)
print('B ∩ A = '+str(con_B_A))
print('===> A ∩ B = B ∩ A  ')
print()
#---------------------------- A ∪ B = B ∪ A ----------------------------
disj_A_B=f_disj(X,A,B)
print('A ∪ B = '+str(disj_A_B))
disj_B_A=f_disj(X,B,A)
print('B ∪ A = '+str(disj_B_A))
print('===> A ∪ B = B ∪ A  ')
print()
#---------------------------- (A ∪ B) ∪ C = A ∪ (B ∪ C) ----------------------------
disj_B_C=f_disj(X,B,C)
print('B ∪ C = '+str(disj_B_C))
disj_AB_C=f_disj(X,disj_A_B,C)
print('(A ∪ B) ∪ C  = '+ str(disj_AB_C))
disj_A_BC=f_disj(X,A,disj_B_C)
print(' A ∪ (B ∪ C) = '+str(disj_A_BC))
print('===> (A ∪ B) ∪ C = A ∪ (B ∪ C)')
print()
#---------------------------- (A ∩ B) ∩ C = A ∩ (B ∩ C) ----------------------------
con_AB_C=f_con(X,con_A_B,C)
print('(A ∩ B) ∩ C  = '+ str(con_AB_C))
con_B_C=f_con(X,B,C)
con_A_BC=f_con(X,A,con_B_C)
print(' A ∩ (B ∩ C) = '+str(con_A_BC))
print('===> (A ∩ B) ∩ C = A ∩ (B ∩ C)')
print()
#---------------------------- A ∪ (B ∩ C) = (A ∪ B) ∩ (A ∪ C) ----------------------------
disj_A_conBC=f_disj(X,A,con_B_C)
print('A ∪ (B ∩ C)       = '+str(disj_A_conBC))
disj_A_C=f_disj(X,A,C)
con_disjAB_disjAC=f_con(X,disj_A_B,disj_A_C)
print('(A ∪ B) ∩ (A ∪ C) = '+str(con_disjAB_disjAC))
print('===> A ∪ (B ∩ C) = (A ∪ B) ∩ (A ∪ C)')
print()
#---------------------------- A ∩ (B ∪ C) =(A ∩ B) ∪ (A ∩ C) ----------------------------
con_A_disjBC=f_con(X,A,disj_B_C)
print('A ∩ (B ∪ C)       = '+str(con_A_disjBC))
con_A_C=f_con(X,A,C)
disj_conAB_conAC=f_disj(X,con_A_B,con_A_C)
print('(A ∩ B) ∪ (A ∩ C) = '+str(disj_conAB_conAC))
print('===> A ∩ (B ∪ C) =(A ∩ B) ∪ (A ∩ C)')
print()
#---------------------------- A ∪ (A ∩ B) = A ----------------------------
disj_A_conAB=f_disj(X,A,con_A_B)
print('A ∪ (A ∩ B) = '+str(disj_A_conAB))
print('A           = '+str(A))
print('===> A ∪ (A ∩ B) = A')
print()
#---------------------------- A ∩ (A ∪ B) = A ----------------------------
con_A_disjAB=f_con(X,A,disj_A_B)
print('A ∩ (A ∪ B) = '+str(con_A_disjAB))
print('A           = '+str(A))
print('===> A ∩ (A ∪ B) = A ')
print()
#---------------------------- A ∪ (¬A ∩ B) = (A ∪ B) ----------------------------
con_notA_B=f_con(X,not_A,B)
disj_A_conNotAB=f_disj(X,A,con_notA_B)
print('A ∪ (¬A ∩ B) = '+str(disj_A_conNotAB))
print('(A ∪ B)      = '+str(disj_A_B))
print('===> A ∪ (¬A ∩ B) = (A ∪ B)')
print()
#---------------------------- A ∩ (¬A ∪ B) = (A ∩ B) ----------------------------
disj_notA_B=f_disj(X,not_A,B)
con_A_disjNotAB=f_con(X,A,disj_notA_B)
print('A ∩ (¬A ∪ B) = '+str(con_A_disjNotAB))
print('(A ∩ B)      = '+str(con_A_B))
print('==> A ∩ (¬A ∪ B) = (A ∩ B)')
print()
#---------------------------- ¬(A ∪ B) = ¬A ∩ ¬B ----------------------------
not_con_AB=f_not_A(X,disj_A_B)
print('¬(A ∪ B) = '+str(not_con_AB))
not_B=f_not_A(X,B)
con_notA_notB=f_con(X,not_A,not_B)
print('¬A ∩ ¬B  = '+str(con_notA_notB))
print('===> ¬(A ∪ B) = ¬A ∩ ¬B')
print()
#---------------------------- ¬(A ∩ B) = ¬A ∪ ¬B ----------------------------
not_con_A_B=f_not_A(X,con_A_B)
print(' ¬(A ∩ B) = '+str(not_con_A_B))
disj_notA_notB=f_disj(X,not_A,not_B)
print(' ¬A ∪ ¬B  = '+str(disj_notA_notB))
print('==> ¬(A ∩ B) = ¬A ∪ ¬B ')
```
