```python
import numpy as np
import matplotlib.pyplot as plt


a=np.arange(0,1.1,0.1)
b=np.arange(0,1.1,0.1)
a,b=np.meshgrid(a,b)

fig=plt.figure(figsize=(8,8))
#-------------------------- minimum -----------------------------
z=a.copy()
for i in range(11):
    for j in range(11):
        if a[i][j]<= b[i][j]:
            z[i][j]=a[i][j]
        else:
            z[i][j]=b[i][j]
ax=fig.add_subplot(221,projection='3d')
wf=ax.plot_wireframe(a,b,z)
ax.set_title('(a) min')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xticks([0,0.5,1])
ax.set_yticks([0,0.5,1])
ax.set_zticks([0,0.5,1])
#-------------------------- algebric product -----------------------------
z=a.copy()
for i in range(11):
    for j in range(11):        
        z[i][j]=a[i][j]*b[i][j]
ax=fig.add_subplot(222,projection='3d')
wf=ax.plot_wireframe(a,b,z)
ax.set_title('(b) algebraic product')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xticks([0,0.5,1])
ax.set_yticks([0,0.5,1])
ax.set_zticks([0,0.5,1])
#-------------------------- bounded product -----------------------------
z=a.copy()
for i in range(11):
    for j in range(11):   
        c=a[i][j]+b[i][j]-1
        if  c >=0 :
            z[i][j]=c
        else:
            z[i][j]=0
ax=fig.add_subplot(223,projection='3d')
wf=ax.plot_wireframe(a,b,z)
ax.set_title('(c) bounded product')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xticks([0,0.5,1])
ax.set_yticks([0,0.5,1])
ax.set_zticks([0,0.5,1])
#-------------------------- drastic product -----------------------------
z=a.copy()
for i in range(11):
    for j in range(11):           
        if  b[i][j] ==1:
            z[i][j]=a[i][j]
        elif a[i][j]==1:
            z[i][j]=b[i][j]
        else:
            z[i][j]=0
ax=fig.add_subplot(224,projection='3d')
wf=ax.plot_wireframe(a,b,z)
ax.set_title('(d) drastic product')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xticks([0,0.5,1])
ax.set_yticks([0,0.5,1])
ax.set_zticks([0,0.5,1])

plt.show()
```
