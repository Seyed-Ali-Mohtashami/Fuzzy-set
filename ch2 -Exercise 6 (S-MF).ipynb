```python
import numpy as np
import matplotlib.pyplot as plt

X=np.arange(0,101,1)

def S(x,l,r):    
    y=np.zeros(x.size)
    for i in range(x.size):
        if x[i]<=l:
            y[i]=0
        elif x[i]<= ((l+r)/2):
            y[i]=(2*(((x[i]-l)/(r-l))**2))
        elif x[i] <=r:
            y[i]=(1-(2*((r-x[i])/(r-l))**2))
        else:
            y[i]=(1)
    return y

left=20
right=50
Y1=S(X,left,right)

left=50
right=80
Y2=S(X,left,right)

plt.xlabel('X')
plt.title('S MF')
plt.text(15,0.6,'l=20,r=50')
plt.text(75,0.6,'l=50,r=80')
plt.plot(X,Y1,X,Y2)
plt.show()
print('crossover point of S(x,l,r) is : x=(l+r)/2')
```
