```python
# ------------------------- step 0 : make a dataset (pairs x1,x2 for inpur and y for output) ----------------------------
import numpy as np
from random import random

s1=-5
s2=5
n=41
x1=np.zeros([n,n])
x2=np.zeros([n,n])
y=np.zeros([n,n])

#print(x1.shape)

for i in range(n):
    for j in range(n):
        x1[i,j]=s1 + (s2-s1)* random()
        x2[i,j]=s1 + (s2-s1)* random()
        y[i,j]=x1[i,j]**2 + x2[i,j]**2

#print(x1)
#print(x2)
#print(y)
```

```python
# ----------------------------------------- step 0 : show x1 and x2 and y) ----------------------------------------------
import matplotlib.pyplot as plt

xx1=np.zeros(n*n)
yx1=np.zeros(n*n)

xx2=np.zeros(n*n)
yx2=np.zeros(n*n)

xy=np.zeros(n*n)
yy=np.zeros(n*n)

for i in range(n):
    for j in range(n):
        xx1[n*i + j]=i
        yx1[i*n + j]=x1[i,j]
        
        xx2[n*i + j]=i
        yx2[i*n + j]=x2[i,j]
        
        xy[n*i + j]=i
        yy[i*n + j]=y[i,j]
#print (xx1.shape)

plt.figure(figsize=(10,10))

plt.subplot(311)
plt.scatter(xx1,yx1,c='red',marker='+')
plt.xticks(np.arange(0,46,5))
plt.yticks(np.arange(-5,+6,1))
plt.title('X1')

plt.subplot(312)
plt.scatter(xx2,yx2,c='blue',marker='+')
plt.xticks(np.arange(0,46,5))
plt.yticks(np.arange(-5,+6,1))
plt.title('X2')

plt.subplot(313)
plt.scatter(yx1,yx2,c='red',marker='o',alpha=0.2)
plt.xticks(np.arange(-5,+6,1))
plt.yticks(np.arange(-5,+6,1))
plt.title('X1 vs X2')

plt.show()
```

```python
rulebase=np.array([[['0000']*3]*7]*7)
print(rulebase.shape)
```

```python
# ----------------------------------------- step 3 : create rulebase table --------------------------------------------
for r in range(7):
    for c in range(7):
        term=''
        index = '0'
        degree='0' 
        for i in range(n):
            for j in range(n):               
                # if antecedent is equal main rule
                if m1[i,j,1] == str(r) and m2[i,j,1]==str(c) :
                        if m3[i,j,3] > degree:
                            term=m3[i,j,0]
                            index=m3[i,j,1]
                            degree=m3[i,j,3]
        rulebase[r,c,0]=str(term)
        rulebase[r,c,1]=str(index)
        rulebase[r,c,2]=str(degree)
print('terms is : ')
print(rulebase[:,:,0])
print()
print('index of terms is : ')
print(rulebase[:,:,1])
print()
print('degree is : ')
print(rulebase[:,:,2])
```

```python
# ------------------------------------------- step 4 : create fuzzy rules ---------------------------------------------
from fuzzy_expert.rule import FuzzyRule
rules=[]

for r in range(7):
    for c in range(7):
        x1_term=x1_terms[r]
        x2_term=x1_terms[c]
        y_term=rulebase[r,c,0]

        rule=FuzzyRule(
                premise=[
                    ('x1',x1_term),
                    ('AND','x2',x2_term)
                ],
                consequence=[
                    ('y',y_term)
                ]
            )
        #print(rule)
        # if x is s2 and y is s3 then z is s2 
        rules.append(rule)

#print(rules)
print(len(rules))
```

```python
# ------------------------ step 5 : test classification (calculate the result of pair inputs) ---------------------------

result_test=np.array([['0000']*n]*n)
print(result_test.shape)
for i in range(n):
    for j in range(n):
        index_term_x1=int(m1[i,j,1])
        index_term_x2=int(m2[i,j,1])
        #print(index_term_x1)
        #print(index_term_x2)
        result_test[i,j]=rulebase[index_term_x1,index_term_x2,0]
        #print('x1 is {} and x2 is {} ---> y is {}',m1[i,j,0],m2[i,j,0],result_test[i,j])
```

```python
# -------------------------------- step 6 : test classification (calculate the errors) ---------------------------------

count_true=0
count_false=0

for i in range(n):
    for j in range (n):
        if m3[i,j,0] == result_test[i,j]:
            count_true+=1
        else:
            count_false+=1
print('count true is : ',str(count_true))
print('count false is : ',str(count_false))
```
