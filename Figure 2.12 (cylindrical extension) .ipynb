```python
import numpy as np
import skfuzzy as fuzzy
import matplotlib.pyplot as plt
#--------------------------------- 2D ---------------------------------
x=np.arange(0,100,1)
y=fuzzy.gaussmf(x,50,20)

fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(221)

#plt.subplot(221)
plt.plot(x,y)
plt.title('(a) Base fuzzy set A')
plt.xlabel('X')
plt.xticks([])
plt.ylabel=('membership grades')

#--------------------------------- 3D ---------------------------------
ax=fig.add_subplot(222,projection='3d')
plt.title('(b) cylindrical extension')
x=np.arange(0,100,1)
y=np.arange(0,10,1)
x , y=np.meshgrid(x,y)
z=fuzzy.gaussmf(x,50,20)


wireframe=ax.plot_wireframe(x,y,z,rstride=1,cstride=5)

ax.view_init(10,70)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('membership grades')
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([0,0.5,1])

plt.show()
```
