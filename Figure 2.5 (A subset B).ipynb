```python
import skfuzzy as fuzzy
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,100,1)
y_a=0.9 * fuzzy.gaussmf(x,60,15)
y_b=fuzzy.gbellmf(x,30,2,50)

plt.plot(x,y_a,x,y_b)

plt.text(40,0.6,'A')
plt.text(20,0.8,'B')

plt.xticks([])

plt.title('A is contained in B')
plt.ylabel('Membership Grades')

plt.show()
```
