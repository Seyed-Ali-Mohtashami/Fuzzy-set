```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzzy

fig=plt.figure(figsize=(15,15))

x=np.arange(-10,11,1.0)
y=np.arange(-10,11,1.0)
x,y=np.meshgrid(x,y)

zx=fuzzy.gbellmf(x,3,3,0)
zy=fuzzy.gbellmf(y,3,3,0)
#z=np.array([[0.0]*zx.size]*zy.size)

#------------------------------------ Zx ------------------------------------
ax=fig.add_subplot(221,projection='3d')
ax.set_xlabel('X')
ax.set_xticks([])
ax.set_ylabel('Y')
ax.set_yticks([])
ax.set_zticks([0,0.5,1])
ax.set_title('Zx')
wireframe=ax.plot_wireframe(x,y,zx,rstride=1,cstride=2)
#------------------------------------ Zy ------------------------------------
ax=fig.add_subplot(222,projection='3d')
ax.set_xlabel('X')
ax.set_xticks([])
ax.set_ylabel('Y')
ax.set_yticks([])
ax.set_zticks([0,0.5,1])
ax.set_title('Zy')
wireframe=ax.plot_wireframe(x,y,zy,rstride=1,cstride=2)
#------------------------------------ min ------------------------------------

for i in range(21):
    for j in range(21):
        if zx[i][j] <= zy[i][j]:
            z[i][j] =zx[i][j]
        elif zx[i][j] > zy[i][j]:
            z[i][j]=zy[i][j]
        
ax=fig.add_subplot(223,projection='3d')
ax.set_xlabel('X')
ax.set_xticks([])
ax.set_ylabel('Y')
ax.set_yticks([])
ax.set_zticks([0,0.5,1])
ax.set_title('(a) z= min (bell(x),bell(y))')
wireframe=ax.plot_wireframe(x,y,z,rstride=1,cstride=1)
#------------------------------------ max ------------------------------------

for i in range(21):
    for j in range(21):
        if zx[i][j] >= zy[i][j]:
            z[i][j] =zx[i][j]
        elif zx[i][j] < zy[i][j]:
            z[i][j]=zy[i][j]
        
ax=fig.add_subplot(224,projection='3d')
ax.set_xlabel('X')
ax.set_xticks([])
ax.set_ylabel('Y')
ax.set_yticks([])
ax.set_zticks([0,0.5,1])
ax.set_title('(a) z= max (bell(x),bell(y))')
wireframe=ax.plot_wireframe(x,y,z,rstride=1,cstride=1)
```
