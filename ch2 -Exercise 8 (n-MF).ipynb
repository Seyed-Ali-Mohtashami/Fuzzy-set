```python
import numpy as np
import matplotlib.pyplot as plt

def S(x,l,r):    
    if x<=l:
        y=0
    elif x<= ((l+r)/2):
        y=(2*(((x-l)/(r-l))**2))
    elif x <=r:
        y=(1-(2*((r-x)/(r-l))**2))
    else:
        y=(1)
    return y
def Z(x,l,r):
    return(1-S(x,l,r))
def n(x,a,c):  
    y=np.zeros(x.size)
    for i in range(x.size):                
        if x[i]<=c:
            y[i]=(S(x[i],c-a,c))
        else:
            y[i]=(Z(x[i],c,c+a))
    return y

X=np.arange(0,101,1)
center=70
spread=20
Y1=n(X,spread,center)
center=40
spread=30
Y2=n(X,spread,center)
plt.title('n MF')
plt.text(10,0.8,'a=30,c=40')
plt.text(80,0.8,'a=20,c=70')
plt.plot(X,Y1,X,Y2)
        
plt.show()   
```
